```python
# ===============================
# Example: Singly Linked List using Excel Data
# ===============================

import pandas as pd

# Step 1: Read the Excel file
file_path = "/content/voleak.xlsx"
df = pd.read_excel(file_path)

# Step 2: Show the first few rows
print("=== Data Preview (head) ===")
print(df.head(), "\n")

# Let's assume your dataset has columns like:
# Student_ID | Name | Score
# If your column names are different, update below
cols = list(df.columns)
print("Detected Columns:", cols, "\n")

# Step 3: Define the Node and Linked List classes
class Node:
    def init(self, data):
        self.data = data
        self.next = None

class SinglyLinkedList:
    def init(self):
        self.head = None

    def append(self, data):
        """Add new node to the end of the list"""
        new_node = Node(data)
        if self.head is None:
            self.head = new_node
            return
        current = self.head
        while current.next:
            current = current.next
        current.next = new_node

    def print_list(self):
        """Print all nodes in the linked list"""
        current = self.head
        print("=== Singly Linked List Data ===")
        while current:
            print(current.data)
            current = current.next
        print()

# Step 4: Convert Excel data rows into Python dictionaries
records = df.to_dict(orient="records")

# Step 5: Create a Singly Linked List and add each record
linked_list = SinglyLinkedList()
for record in records:
    linked_list.append(record)

# Step 6: Display the linked list
linked_list.print_list()
```

```python
import pandas as pd

# Step 1: Load the dataset
df = pd.read_excel("/content/voleak.xlsx")

# Step 2: Show first few rows
print("ðŸ“˜ Data Preview:")
print(df.head(), "\n")

# Step 3: Convert first column to a list (e.g., student IDs or Names)
students = df.iloc[:, 0].tolist()  # You can replace with df['Name'].tolist() if column exists
print("ðŸŽ¯ Sample Students List:")
print(students[:10], "\n")

# Step 4: Define Node class
class Node:
    def init(self, data):
        self.data = data
        self.next = None

# Step 5: Define Circular Linked List class
class CircularLinkedList:
    def init(self):
        self.head = None

    def append(self, data):
        new_node = Node(data)
        if not self.head:
            self.head = new_node
            self.head.next = self.head
        else:
            temp = self.head
            while temp.next != self.head:
                temp = temp.next
            temp.next = new_node
            new_node.next = self.head

    def display(self):
        nodes = []
        if not self.head:
            return nodes
        temp = self.head
        while True:
            nodes.append(temp.data)
            temp = temp.next
            if temp == self.head:
                break
        return nodes

    def delete(self, key):
        if not self.head:
            return
        curr = self.head
        prev = None

        # Case 1: Head node deletion
        if curr.data == key:
            while curr.next != self.head:
                curr = curr.next
            curr.next = self.head.next
            self.head = self.head.next
            return

        # Case 2: Delete non-head node
        curr = self.head
        while curr.next != self.head:
            prev = curr
            curr = curr.next
            if curr.data == key:
                prev.next = curr.next
                return

# Step 6: Create and display Circular Linked List
cll = CircularLinkedList()
for student in students[:10]:  # limit for clarity
    cll.append(student)

# Step 7: Display the list
print("ðŸ” Circular Linked List (First 10 Students):")
print(" -> ".join(map(str, cll.display())), "\n")

# Step 8: Delete one element
if students:
    to_delete = students[3]  # Delete 4th student
    cll.delete(to_delete)
    print(f"ðŸ—‘ After deleting '{to_delete}':")
    print(" -> ".join(map(str, cll.display())))
```
